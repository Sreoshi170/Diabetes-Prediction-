```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe().T
```

```python
df["Outcome"].value_counts()
```

```python
df.groupby("Outcome").mean()
```

```python
X=df.iloc[:,:-1]    #X=df.drop(columns='Outcome',axis=1)
y=df["Outcome"]
```

```python
print(X)
```

```python
print(y)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
ss=scaler.fit_transform(X)
```

```python
print(ss)
```

```python
X=ss
y=df['Outcome']
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split    #train-test-split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#training the model
from sklearn.svm import SVC           #SUPPORT VECTOR MACHINE
classifier=SVC(kernel='linear')
```

```python
#trainig the support vector machine 
classifier.fit(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression    #LOGISTICS REGRESSION
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor       #DECISION TREE
dt=DecisionTreeRegressor()
```

```python
dt.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor      #RANDOM FOREST
rf=RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsRegressor       #KNN
knn=KNeighborsRegressor()
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB        #NAIVE BAYES
nb=GaussianNB()
```

```python
nb.fit(X_train,y_train)
```

```python
#evaluation
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score      #CLASSIFICATION REPORT
```

```python
y_pred_train=classifier.predict(X_train)     #TRAINING DATA ACCURACY SCORE(SUPPORT VECTOR MACHINE)
training_data_accuracy=accuracy_score(y_pred_train,y_train)   #ACCURACY SCORE
training_data_precision=precision_score(y_pred_train,y_train)  #PRECISION SCORE
training_data_recall=recall_score(y_pred_train,y_train)   #RECALL
training_data_f1_score=f1_score(y_pred_train,y_train)   #F1-SCORE
print("For Training Data:")
print("Accuracy:",training_data_accuracy)
print("Precision:",training_data_precision)
print("Recall:",training_data_recall)
print("F1-Score:",training_data_f1_score)
```

```python
y_pred_test=classifier.predict(X_test)     #TESTNING DATA ACCURACY SCORE(SUPPORT VECTOR MACHINE)
testing_data_accuracy=accuracy_score(y_pred_test,y_test)   #ACCURACY SCORE
testing_data_precision=precision_score(y_pred_test,y_test)  #PRECISION SCORE
testing_data_recall=recall_score(y_pred_test,y_test)      #RECALL 
testing_data_f1_score=f1_score(y_pred_test,y_test)     #F1-SCORE
print("For Testing Data:")
print("Accuracy:",testing_data_accuracy) 
print("Precision:",testing_data_precision)
print("Recall:",testing_data_recall)
print("F1-Score:",testing_data_f1_score)
```

```python
y_pred_train=lr.predict(X_train)     #TRAINING DATA ACCURACY SCORE(LOGISTICS REGRESSION)
training_data_accuracy=accuracy_score(y_pred_train,y_train)   #ACCURACY SCORE
training_data_precision=precision_score(y_pred_train,y_train)  #PRECISION SCORE
training_data_recall=recall_score(y_pred_train,y_train)   #RECALL
training_data_f1_score=f1_score(y_pred_train,y_train)   #F1-SCORE
print("For Training Data:")
print("Accuracy:",training_data_accuracy)
print("Precision:",training_data_precision)
print("Recall:",training_data_recall)
print("F1-Score:",training_data_f1_score)
```

```python
y_pred_test=lr.predict(X_test)     #TESTNING DATA ACCURACY SCORE(LOGISTICS REGRESSION)
testing_data_accuracy=accuracy_score(y_pred_test,y_test)   #ACCURACY SCORE
testing_data_precision=precision_score(y_pred_test,y_test)  #PRECISION SCORE
testing_data_recall=recall_score(y_pred_test,y_test)      #RECALL 
testing_data_f1_score=f1_score(y_pred_test,y_test)     #F1-SCORE
print("For Testing Data:")
print("Accuracy:",testing_data_accuracy) 
print("Precision:",testing_data_precision)
print("Recall:",testing_data_recall)
print("F1-Score:",testing_data_f1_score)
```

```python
y_pred_train=dt.predict(X_train)     #TRAINING DATA ACCURACY SCORE(DECISION TREE)
training_data_accuracy=accuracy_score(y_pred_train,y_train)   #ACCURACY SCORE
training_data_precision=precision_score(y_pred_train,y_train)  #PRECISION SCORE
training_data_recall=recall_score(y_pred_train,y_train)   #RECALL
training_data_f1_score=f1_score(y_pred_train,y_train)   #F1-SCORE
print("For Training Data:")
print("Accuracy:",training_data_accuracy)
print("Precision:",training_data_precision)
print("Recall:",training_data_recall)
print("F1-Score:",training_data_f1_score)
```
